###  Análisis Univariado

```python
df.dtypes
```

```python
# Histograma de la variable 'HP'
sns.histplot(df['base_friendship'], kde=True)
plt.title('Distribución de HP')
plt.show()

# Diagrama de caja para 'Attack'
sns.boxplot(x=df['percent_female'])
plt.title('Diagrama de Caja de Attack')
plt.show()
```

```python
df['base_friendship']=df['base_friendship'].astype('int64')
```

### Análisis Bivariado
*    Relación entre dos variables
*    Diagramas de dispersión y mapas de calor

```python
# Diagrama de dispersión entre 'Attack' y 'Defense'
sns.scatterplot(x='attack', y='defense', data=df)
plt.title('Attack vs Defense')
plt.show()

# Mapa de calor de correlación
SinNombre=df.select_dtypes(include=np.number)
corr = SinNombre.corr()
plt.figure(figsize=(12, 12))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlación')
plt.show()
```

```python
SinNombre['Suma De estadisticas']=SinNombre['hp']+SinNombre['attack']+SinNombre['defense']+SinNombre['sp_atk']+SinNombre['sp_def']+SinNombre['speed']
```

```python
df.drop(columns=['total','height','catch_rate'], inplace=True)
```

### Análisis Multivariado
*    Relación entre múltiples variables
*    Pairplot y agrupaciones

```python
df['type1'].value_counts()
```

```python
# Pairplot de las estadísticas principales
SoloAlgunosTipos=df[df['type1'].isin(['Rock', 'Steel', 'Water','Fairy', 'Planeta'])]

sns.pairplot(SoloAlgunosTipos[['hp', 'attack', 'defense', 'weight', 'speed', 'type1']], hue='type1')
plt.show()
```

```python
# Agrupación por tipo de Pokémon y estadísticas promedio
df.groupby('type1').mean().sort_values('Total', ascending=False)
```

### Visualización de Datos
*    Gráficos avanzados para insights más profundos

```python
# Boxplot de estadísticas por tipo de Pokémon
plt.figure(figsize=(12, 6))
sns.boxplot(x='type1', y='defense', data=df)
plt.xticks(rotation=90)
plt.title('Total Stats por Tipo de Pokémon')
plt.show()
```
