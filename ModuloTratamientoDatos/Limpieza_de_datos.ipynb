Carga de librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV
df = pd.read_csv('sample_data/Titanic.csv')
df.head()
```

### Identificación de Valores Nulos

```python
# Identificar valores nulos
df.isnull().sum()

# Visualización de valores nulos
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='viridis')
plt.title('Valores Nulos en el Conjunto de Datos del Titanic')
plt.show()
```

### Tratamiento de Valores Nulos

```python
# Eliminando Cabin
df.drop(columns=['Cabin'], inplace=True)

# Eliminar filas con valores nulos en la columna 'Age'
df.dropna(subset=['Age'], inplace=True)
#df['Age'].fillna(df['Age'].mean(), inplace=True)

# Imputar valores nulos en la columna 'Embarked' con el valor más frecuente
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Visualización de valores nulos
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cmap='viridis')
plt.title('Valores Nulos en el Conjunto de Datos del Titanic')
plt.show()
```

### Detección y Tratamiento de Valores Atípicos

```python
# Boxplot de la columna 'Fare2' para identificar valores atípicos
plt.figure(figsize=(8, 6))
sns.boxplot(x='Fare', data=df)
plt.title('Diagrama de Caja de la Tarifa')
plt.show()

# Tratar valores atípicos (ejemplo: truncar en el percentil 95)
fare_90 = df['Fare'].quantile(0.90)
df['Fare'] = df['Fare'].apply(lambda x: fare_90 if x > fare_90 else x)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Fare', data=df)
plt.show()
```

Codificación de Variables Categóricas

```python
# Codificación one-hot para la columna 'Sex'
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)

# Codificación de etiqueta para la columna 'Embarked'     .list
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

Visualización Post-Limpieza

```python
# Histograma de la edad después de la limpieza
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribución de Edad después de la Limpieza')
plt.show()
```

```python
sns.pairplot(df, corner=True, hue='Survived')
plt.show()
```
