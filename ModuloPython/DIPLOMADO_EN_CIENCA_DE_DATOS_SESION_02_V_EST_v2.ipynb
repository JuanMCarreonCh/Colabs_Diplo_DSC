Antes de comenzar, es importante entender qué es el comando `print` en Python. El comando `print` se utiliza para mostrar información en la consola o en la salida estándar del programa. Cuando utilizas `print`, le estás indicando a Python que muestre en la pantalla el valor o los valores que le pasas como argumento.

Es crucial notar que cuando utilizas `print`, debes rodear el texto que deseas mostrar con comillas. Por ejemplo, si escribes `print("Hola, mundo!")`, Python mostrará en la consola la frase "Hola, mundo!". Estas comillas (dobles o simples) son esenciales ya que le indican a Python que el texto entre ellas debe ser tratado como una cadena de caracteres (string -> str).

Por ejemplo, si intentas mostrar texto sin comillas como en `print(Hola, mundo!)`, Python mostrará un error porque no sabe qué es `Hola` y `mundo`, pero si utilizas comillas, como en `print("Hola, mundo!")`, Python entenderá que es un texto que debe ser mostrado en la consola.

Así que, cuando escribes `print`, asegúrate de rodear el texto que deseas mostrar con comillas para que Python lo interprete correctamente. Es una forma útil de ver lo que está sucediendo en tu programa mientras se ejecuta.

### ¿Qué es NumPy?

NumPy es una biblioteca de Python que proporciona una amplia gama de herramientas y funciones para trabajar con matrices y arreglos multidimensionales. Es una de las bibliotecas más utilizadas en el campo de la computación científica y la manipulación de datos numéricos en Python.

### Ventajas de usar NumPy:

1. **Eficiencia en el manejo de datos**: NumPy está escrito en lenguaje C, lo que le otorga una alta eficiencia en la manipulación de datos numéricos. Las operaciones con NumPy suelen ser mucho más rápidas que las operaciones equivalentes utilizando los tipos de datos nativos de Python.

2. **Soporte para operaciones vectorizadas**: NumPy permite realizar operaciones vectorizadas, lo que significa que puedes realizar operaciones en matrices y arreglos completos de datos de una sola vez, sin necesidad de utilizar bucles explícitos en Python. Esto hace que el código sea más conciso y eficiente.

3. **Amplia gama de funciones matemáticas y estadísticas**: NumPy incluye una gran cantidad de funciones matemáticas y estadísticas integradas, como suma, resta, multiplicación, división, potenciación, raíz cuadrada, funciones trigonométricas, funciones exponenciales, entre otras. Esto facilita realizar cálculos numéricos complejos de manera rápida y sencilla.

4. **Manipulación de datos de forma eficiente**: NumPy proporciona herramientas para manipular y transformar datos de manera eficiente, incluyendo funciones para transponer matrices, redimensionar arreglos, dividir y combinar datos, entre otras operaciones.

5. **Integración con otras bibliotecas**: NumPy se integra fácilmente con otras bibliotecas de Python, como SciPy (para computación científica), Matplotlib (para visualización de datos), Pandas (para análisis de datos), y más. Esto te permite construir soluciones completas para tus necesidades de análisis y procesamiento de datos.

En resumen, NumPy es una biblioteca poderosa y versátil que facilita la manipulación y el análisis de datos numéricos en Python, gracias a su eficiencia, amplia gama de funciones y capacidad de integración con otras herramientas y bibliotecas.

### Tipos Integer (`int`)

Los enteros (`int`) son un tipo de dato en Python que representa números enteros, es decir, números sin parte fraccionaria. Los enteros pueden ser positivos, negativos o cero. En Python, los enteros se representan utilizando el tipo de datos `int`.

Por ejemplo, `5`, `-3`, y `0` son ejemplos de enteros en Python. Puedes realizar operaciones matemáticas comunes con enteros, como suma, resta, multiplicación y división entera.

### Tipos Float (`float`)

Los números de punto flotante (`float`) son otro tipo de dato numérico en Python que representa números con parte fraccionaria. Estos números pueden ser representados con una parte decimal y/o una parte exponencial. En Python, los números de punto flotante se representan utilizando el tipo de datos `float`.

Por ejemplo, `3.14`, `-0.001`, y `2.0` son ejemplos de números de punto flotante en Python. Puedes realizar operaciones matemáticas comunes con números de punto flotante, incluyendo suma, resta, multiplicación, división y operaciones más avanzadas como potenciación y raíz cuadrada.

El operador de comparación en Python, también conocido como operador relacional, se utiliza para comparar dos valores y determinar la relación entre ellos. Estos operadores devuelven un valor booleano (`True` o `False`) según si la comparación es verdadera o falsa.

Los operadores de comparación disponibles en Python son:

- `==`: Igual que. Retorna `True` si los operandos son iguales.
- `!=`: Distinto de. Retorna `True` si los operandos son diferentes.
- `<`: Menor que. Retorna `True` si el operando de la izquierda es menor que el de la derecha.
- `>`: Mayor que. Retorna `True` si el operando de la izquierda es mayor que el de la derecha.
- `<=`: Menor o igual que. Retorna `True` si el operando de la izquierda es menor o igual que el de la derecha.
- `>=`: Mayor o igual que. Retorna `True` si el operando de la izquierda es mayor o igual que el de la derecha.

Ejemplos:

````python
# Definir dos variables
x = 5
y = 10

# Utilizar operadores de comparación
print("x == y:", x == y)  # False, ya que 5 no es igual a 10
print("x != y:", x != y)  # True, ya que 5 es diferente de 10
print("x < y:", x < y)    # True, ya que 5 es menor que 10
print("x > y:", x > y)    # False, ya que 5 no es mayor que 10
print("x <= y:", x <= y)  # True, ya que 5 es menor o igual que 10
print("x >= y:", x >= y)  # False, ya que 5 no es mayor o igual que 10

La sintaxis básica de un ciclo for es la siguiente:


image.png

* El bloque de código indentado que sigue al encabezado del ciclo `for` se ejecutará en cada iteración, con `elemento` tomando el valor del siguiente elemento en la secuencia en cada iteración.


* Ejemplos:

# Bucle `while` en Python
El bucle `while` en Python se utiliza para repetir un bloque de código mientras se cumpla una condición específica. Es útil cuando no sabemos cuántas veces necesitamos repetir un conjunto de instrucciones, pero sabemos que queremos continuar hasta que se cumpla cierta condición.

image.png



La sintaxis básica de un bucle `while` es la siguiente:

```python
while condicion:
    # Bloque de código a ejecutar mientras se cumpla la condición

```

* La palabra clave `while` indica el inicio del bucle.
* `condicion` es una expresión que se evalúa como verdadera o falsa. Mientras la condición sea verdadera, el bloque de código dentro del bucle while se ejecutará repetidamente.
* El bloque de código indentado que sigue al encabezado del bucle `while` se ejecutará en cada iteración mientras la condición sea verdadera.

Es importante tener cuidado al usar un bucle while, ya que si la condición nunca se vuelve falsa, el bucle puede ejecutarse infinitamente, lo que puede provocar que el programa se bloquee.

Ejemplo:

`````python
again = 'sí'
while again == 'sí':
  print("¡Hola!")
  again = input("¿Quieres que se repita?: ")
`````

`````python
total = 10
while total < 100:
  num = int(input("Introduce un número: "))
  total = total + num
  print("El total es: ", total)
`````

Los operadores de comparación disponibles en Python son:

- `==`: Igual que. Retorna `True` si los operandos son iguales.
- `!=`: Distinto de. Retorna `True` si los operandos son diferentes.
- `<`: Menor que. Retorna `True` si el operando de la izquierda es menor que el de la derecha.
- `>`: Mayor que. Retorna `True` si el operando de la izquierda es mayor que el de la derecha.
- `<=`: Menor o igual que. Retorna `True` si el operando de la izquierda es menor o igual que el de la derecha.
- `>=`: Mayor o igual que. Retorna `True` si el operando de la izquierda es mayor o igual que el de la derecha.

Ejemplos:

```python
# Definir dos variables
x = 5
y = 10

# Utilizar operadores de comparación
print("x == y:", x == y)  # False, ya que 5 no es igual a 10
print("x != y:", x != y)  # True, ya que 5 es diferente de 10
print("x < y:", x < y)    # True, ya que 5 es menor que 10
print("x > y:", x > y)    # False, ya que 5 no es mayor que 10
print("x <= y:", x <= y)  # True, ya que 5 es menor o igual que 10
print("x >= y:", x >= y)  # False, ya que 5 no es mayor o igual que 10
```

# Tuplas, listas y diccionarios.

## Listas
* Descripción: Una lista es una estructura de datos mutable que puede contener una colección ordenada de elementos de diferentes tipos.
* Sintaxis: Se define utilizando corchetes [] y los elementos están separados por comas.
Ejemplo:
```python
mi_lista = [1, 'hola', True, 3.14]
```

## Tuplas
* Descripción: Una tupla es una estructura de datos inmutable que puede contener una colección ordenada de elementos de diferentes tipos.
* Sintaxis: Se define utilizando paréntesis () y los elementos están separados por comas.
Ejemplo:
```python
mi_tupla = (1, 'hola', True, 3.14)
```

## Diccionarios
* Descripción: Un diccionario es una estructura de datos mutable que contiene una colección desordenada de pares clave-valor, donde cada clave está asociada con un valor.
* Sintaxis: Se define utilizando llaves {} y los pares clave-valor están separados por comas, con la clave y el valor separados por :.
Ejemplo:
```python
mi_diccionario = {'nombre': 'Juan', 'edad': 30, 'ciudad': 'México'}
```

## Similitudes y diferencias.


| Característica | Listas                  | Tuplas               | Diccionarios              |
|----------------|-------------------------|----------------------|---------------------------|
| Mutabilidad    | Mutable                 | Inmutable            | Mutable                   |
| Sintaxis       | `[]`                     | `()`                  | `{}`                      |
| Orden          | Ordenadas               | Ordenadas            | Desordenadas              |
| Elementos      | Pueden repetirse        | Pueden repetirse     | Las claves deben ser únicas |
| Acceso         | Índices (inicio en 0)   | Índices (inicio en 0)| Claves                    |

En Python, hay dos formas principales de ordenar elementos en una lista: el método `sort()` y la función `sorted()`. Aunque ambos se utilizan para ordenar elementos, tienen diferencias importantes en su funcionamiento y en cómo afectan a la lista original. Aquí están las diferencias clave:

1. **Método `sort()`:**
   - El método `sort()` se aplica directamente a la lista y modifica la lista original en su lugar.
   - Ordena los elementos de la lista en orden ascendente por defecto, pero también puede tomar un argumento `reverse=True` para ordenar en orden descendente.
   - No devuelve una nueva lista ordenada, sino que modifica la lista original.
   - Es útil cuando se quiere modificar la lista original y no es necesario conservar la versión original de la lista.

2. **Función `sorted()`:**
   - La función `sorted()` toma una lista como argumento y devuelve una nueva lista ordenada, sin modificar la lista original.
   - Por defecto, ordena los elementos en orden ascendente, pero también puede tomar un argumento `reverse=True` para ordenar en orden descendente.
   - Es útil cuando se quiere mantener la lista original sin cambios y obtener una nueva lista ordenada.

Aquí hay un ejemplo que ilustra la diferencia entre el método `sort()` y la función `sorted()`:

```python
# Lista original
numeros = [5, 2, 7, 1, 3]

# Método sort(): Modifica la lista original en su lugar
numeros.sort()
print("Lista ordenada con método sort():", numeros)  # Output: [1, 2, 3, 5, 7]

numeros = [5, 2, 7, 1, 3, 31, 1, 5, 3, 62]
# Función sorted(): Devuelve una nueva lista ordenada sin modificar la lista original
numeros_ordenados = sorted(numeros)
print("Nueva lista ordenada con función sorted():", numeros_ordenados)  # Output: [1, 2, 3, 5, 7]
print("Lista original sin cambios:", numeros)  # Output: [5, 2, 7, 1

# Diccionarios.
Los diccionarios en Python son una estructura de datos que permite almacenar una colección de pares clave-valor. Cada elemento en un diccionario está compuesto por una clave única y su correspondiente valor. La principal característica de los diccionarios es que permiten acceder rápidamente a los valores asociados a una clave dada.

## Sintaxis
Los diccionarios en Python se definen utilizando llaves {}. Cada par clave-valor se separa por dos puntos : y los elementos se separan por comas. La sintaxis general es la siguiente:


```python
diccionary = {clave1: valor1, clave2: valor2, clave3: valor3}
```
## Características
* Los diccionarios son mutables, lo que significa que pueden modificarse después de su creación.
* Las claves en un diccionario deben ser únicas y pueden ser de cualquier tipo inmutable, como enteros, cadenas o tuplas.
* Los valores en un diccionario pueden ser de cualquier tipo, incluyendo otros diccionarios, listas o incluso funciones.

## Operaciones comunes
* Agregar elementos: Se puede agregar un nuevo par clave-valor al diccionario utilizando la sintaxis mi_diccionario[nueva_clave] = nuevo_valor.
* Acceder a elementos: Se puede acceder al valor asociado a una clave específica utilizando la sintaxis mi_diccionario[clave].
* Modificar elementos: Se puede modificar el valor asociado a una clave existente utilizando la sintaxis mi_diccionario[clave] = nuevo_valor.
* Eliminar elementos: Se puede eliminar un par clave-valor del diccionario utilizando la palabra clave del seguida de la clave a eliminar, por ejemplo del mi_diccionario[clave].
* Obtener todas las claves o valores: Se pueden obtener todas las claves o valores del diccionario utilizando los métodos keys() o values(), respectivamente.

Los diccionarios son una herramienta poderosa en Python que se utiliza ampliamente en una variedad de aplicaciones, desde el manejo de datos hasta la implementación de algoritmos.

### JSON (JavaScript Object Notation)

JSON (JavaScript Object Notation) es un formato de intercambio de datos ligero y fácil de leer. Se utiliza comúnmente para enviar datos entre un servidor y un navegador web, pero también es ampliamente utilizado en el intercambio de datos entre diferentes sistemas y aplicaciones.

JSON se basa en dos estructuras de datos principales: objetos y arrays. Un objeto JSON es similar a un diccionario en Python, con pares clave-valor separados por comas y encerrados entre llaves `{}`. P. ej.:

}
